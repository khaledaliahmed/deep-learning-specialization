## `seaborn`

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import os
```

### Load the Dataset

`embeded Datasets in seaborn`

```python
## load the dataset, which is embeded in seaborn
df_tips = sns.load_dataset('tips')

df_tips.head()
```

### `Countplot - barplot`

```python
## Let's get the countplot of (time) variable

plt.figure(figsize=(10, 6))    ## figure shape

sns.countplot(x='time', data=df_tips)

## title and axis
plt.title('Distr. of time variable in Dataset', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.show()
```

```python
## Going Further

plt.figure(figsize=(10, 6))    ## figure shape

sns.countplot(x='time', data=df_tips, hue='sex')

## title and axis
plt.title('Distr. of time variable in Dataset', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.legend(loc='upper left')
plt.show()
```

```python
## for countplot --> you can use  --> sns.catplot(kind='count')
```

### `Boxplot`

```python
## what about boxplot for (total_bill) column
plt.figure(figsize=(10, 6))    ## figure shape

sns.boxplot(y='total_bill', data=df_tips)

## title and axis
plt.title('Distr of total bill', fontsize=14)
plt.xlabel('Total bill', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.show()
```

```python
## what about going further by using catplot with (sex) variable
plt.figure(figsize=(12, 8))    ## figure shape

sns.catplot(y='total_bill', x='sex', data=df_tips, kind='box')

## title and axis
plt.title('Distr of total bill', fontsize=14)
plt.xlabel('Sex', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.show()
```

```python
## what about going further by using catplot with (time) variable
plt.figure(figsize=(12, 8))    ## figure shape

sns.catplot(y='total_bill', x='time', data=df_tips, kind='box')

## title and axis
plt.title('Distr of total bill', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.show()
```

```python
## for boxplot you can use --> sns.catplot(kind='box')
```

### `Scatter plot`

```python
## What about scatter plot

plt.figure(figsize=(10, 6))    ## figure shape

sns.scatterplot(x='total_bill', y='tip', data=df_tips)

## title and axis
plt.title('Relation between total bill and tip', fontsize=14)
plt.xlabel('Total bill', fontsize=14)
plt.ylabel('Tip', fontsize=14)

plt.show()
```

```python
## Going Further

plt.figure(figsize=(10, 6))    ## figure shape

sns.scatterplot(x='total_bill', y='tip', data=df_tips, hue='sex')

## title and axis
plt.title('Relation between total bill and tip', fontsize=14)
plt.xlabel('Total bill', fontsize=14)
plt.ylabel('Tip', fontsize=14)

plt.legend(loc='upper left')
plt.show()
```

### `Subplots - replot`

```python
## What about subplots using replot function in seaborn with (smoker) variable as two columns
plt.figure(figsize=(14, 8))    ## figure shape

sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='smoker');  ## as two columns
```

```python
## What about subplots using replot function in seaborn with (smoker) variable as two rows
plt.figure(figsize=(14, 8))    ## figure shape

sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', row='smoker');  ## as two rows
```

```python
## What about subplots using replot function in seaborn with (smoker & time) variables
plt.figure(figsize=(10, 6))    ## figure shape
## as two rows and two columns
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='smoker', row='time'); 
```

```python
## What about subplots using replot function in seaborn with (size) variable as (hue & size)
plt.figure(figsize=(10, 6))    ## figure shape

## as two rows and two columns
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', col='smoker', row='time', size='size', hue='size');  
```

```python
## What about subplots using replot function in seaborn with (smoker) variable as (style and hue) and alpha (opacity)
plt.figure(figsize=(10, 6))    ## figure shape

## as two rows and two columns
sns.relplot(x='total_bill', y='tip', data=df_tips, kind='scatter', style='smoker', hue='smoker', alpha=0.4); 

## title and axis
plt.title('Relation between total bill and tips with smoker', fontsize=14)
plt.xlabel('Total bill', fontsize=14)
plt.ylabel('Tip', fontsize=14)

plt.show()
```

### `Histogram` 

```python
## Let's make a Histogram
plt.figure(figsize=(10, 6))    ## figure shape

sns.histplot(x='total_bill', data=df_tips, bins=50, kde=True, hue='sex')  ## I add hue=sex --> multi histogram

## title and axis
plt.title('Distr. of total bill in Dataset', fontsize=14)
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

### `Line plot - replot`

```python
## Getting the path of Dataset and read the CSV File using pandas
path = os.path.join(os.getcwd(), 'files', 'AirQuality.csv')
df_air = pd.read_csv(path, sep=';')  ## take care of the separator
df_air.head()
```

```python
## get the DataFrame where there is no null --< not that (~) this mark reverse the condition
df_air = df_air[~df_air['Time'].isna()]


## convert the Time column to integer then append to the list
time_lst = []
for i in range(len(df_air)):
    time_lst.append(df_air['Time'][i][:2])
    
## override the Time column
df_air['Time'] = time_lst

df_air.head()
```

```python
## what about showing the disribution of (NO2-GT) with time
plt.figure(figsize=(14, 8))    ## figure shape

sns.relplot(x='Time', y='NO2(GT)', data=df_air, kind='line', markers=True)

## 
plt.title('Relation between Time and NO2', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.ylabel('NO2(GT)', fontsize=14)
plt.xticks(rotation=90)
plt.show()
```
