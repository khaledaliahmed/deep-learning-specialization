## `matplotlib`

> `Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.`

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

## %matplotlib qt   ---> to run the output in outer interface


import os
```

```python
## you can use different styles for plotting using matplotlib

plt.style.use('default')

# plt.style.use('ggplot')
```

[Check more styles](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

### `Load the Dataset`

```python
## get the housing dataset
path = os.path.join(os.getcwd(), 'files', 'housing.csv')

df_housing = pd.read_csv(path)

## check the head of this DF
df_housing.head()
```

### `ScatterPlot`

```python
## Create a scatter plot between two numerical variables

plt.scatter(df_housing['longitude'], df_housing['latitude'])

## name the axis
plt.xlabel('Longitude')
plt.ylabel('Latitude')

## title of the plot
plt.title('Longitude vs. Latitude of Housing Dataset')

## finally show the graph
plt.show()
```

```python
## what about more options
## change the color, opacity, marker, figure size, and more

plt.figure(figsize=(10, 6))  ## change the shape of figure

plt.scatter(x=df_housing['longitude'], y=df_housing['latitude'], marker='o', color='black', 
            cmap=plt.get_cmap('jet'))

## name the axis
plt.xlabel('Longitude', fontsize=14, color='red')
plt.ylabel('Latitude', fontsize=14, color='red')

## title of the plot
plt.title(label='Longitude vs. Latitude of Housing Dataset', fontsize=15, color='red')

## finally show the graph
plt.show()
```

```python
## what about more options
## change the color, opacity, marker, figure size, color bar, 

plt.figure(figsize=(14, 8))  ## change the shape of figure

sc = plt.scatter(x=df_housing['longitude'], y=df_housing['latitude'], marker='o', c=df_housing['median_house_value'], 
            cmap=plt.get_cmap('jet'), alpha=0.4, s=df_housing['population']/100, label='Population')

plt.colorbar(sc)

## name the axis
plt.xlabel('Longitude', fontsize=14, color='k')
plt.ylabel('Latitude', fontsize=14, color='k')

## title of the plot
plt.title(label='Longitude vs. Latitude of Housing Dataset', fontsize=15, color='k')

## show the legend
plt.legend()

## finally show the graph
plt.show()
```

```python
## we can create the same plot exactly using pandas

df_housing.plot(kind='scatter', x='longitude', y='latitude', c='median_house_value', cmap=plt.get_cmap('jet'), alpha=0.4,
                s=df_housing['population']/100, label='population', figsize=(14, 8), colorbar=True,
                xlabel='Longitude', ylabel='Latitude', title='Longitude vs. Latitude of Housing Dataset')

## finally show the graph
plt.show()
```

```python
### another scatter plot

plt.figure(figsize=(10, 6))  ## figure shape

x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.rand(100, 1)
z = 4 - 3 * x + np.random.rand(100, 1)

plt.scatter(x, y, color='r', label='x & y')
plt.scatter(x, z, color='g', label='x & z')


## title and axis
plt.title('Scatterplot for some test data', fontsize=12)
plt.xlabel('x', fontsize=12)

## legend
plt.legend()
plt.show()
```

### `LinePlot`

```python
## creat a test dataset
x = np.linspace(1, 5, 100) 
y = np.exp(x)  
z = np.log2(x)
```

```python
## plot as line
plt.plot(x, y)

## name the axis and make a title
plt.title('x vs. y')

plt.xlabel('x')
plt.ylabel('y')

## show the graph
plt.show()
```

```python
## Go Further

## plot as line
plt.plot(x, y, marker='o', color='r', markersize=2, label='exp(x)')
plt.plot(x, z, marker='+', color='g', markersize=2, label='log10(x)')

## name the axis and make a title
plt.title('x vs. y')

plt.xlabel('x')
plt.ylabel('y')

## legend
plt.legend()

## show the graph
plt.show()
```

```python
## twinx

## We want to optimize the above figure
fig, ax = plt.subplots()

ax.plot(x, y, color='blue', label='exp(x)')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', color='blue', fontsize=14)

## create another axis
ax2 = ax.twinx()
ax2.plot(x, z, color='red', label='log10(x)')

ax2.set_ylabel('z', color='red', fontsize=14)

plt.title('x vs. y & x vs. z', fontsize=14)
plt.show()
```

```python
## what about some annotations

## plot as line
plt.plot(x, y)

## name the axis and make a title
plt.title('x vs. y')

plt.xlabel('x')
plt.ylabel('y')

## Annotations
plt.annotate('max value', xy=(5, 140), xytext=(4, 20), 
             arrowprops={"arrowstyle":"->","color":"red"})

## show the graph
plt.show()
```

### `Histogram`

### `Subplots`
`what about suplots` <br />
`Let say we want to get the histogram for (median_house_value) at each ocean_proximity alone.`

```python
## check
df_housing['ocean_proximity'].value_counts()
```

```python
## example
df_housing[df_housing['ocean_proximity']=='INLAND'].head()
```

```python
## create a suplots, I have 4 main categories, let's make 2 rows x 2 columns

fig, ax = plt.subplots(2, 2, figsize=(20, 10))
fig.tight_layout(pad=5)

plt.sca(ax[0, 0])
to_plot = df_housing[df_housing['ocean_proximity']=='<1H OCEAN']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=<1H OCEAN', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.sca(ax[0, 1])
to_plot = df_housing[df_housing['ocean_proximity']=='INLAND']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=INLAND', fontsize=14)


plt.sca(ax[1, 0])
to_plot = df_housing[df_housing['ocean_proximity']=='NEAR OCEAN']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=NEAR OCEAN', fontsize=14)
plt.xlabel('Houses Prices', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.sca(ax[1, 1])
to_plot = df_housing[df_housing['ocean_proximity']=='NEAR BAY']['median_house_value']
plt.hist(to_plot, bins=50)
plt.title('Distr. of Houses Prices at ocean_proximity=NEAR BAY', fontsize=14)
plt.xlabel('Houses Prices', fontsize=14)


plt.show()
```

```python
## optimize the above code to be more professional
## create a suplots, I have 4 main categories, let's make 2 rows x 2 columns

fig, ax = plt.subplots(2, 2, figsize=(20, 10))
fig.tight_layout(pad=5)

## unique values for (ocean_proximity) except one value (which is very small and useless)
## read this code carefully to understand it -- trace the code
ocean_vals = df_housing['ocean_proximity'].value_counts().index[:4].tolist()
ocean_vals = np.array(ocean_vals).reshape(2, 2)


for row in range(2):
    for col in range(2):
        plt.sca(ax[row, col])
        to_plot = df_housing[df_housing['ocean_proximity']==ocean_vals[row, col]]['median_house_value']
        plt.hist(to_plot, bins=50)
        plt.title(f'Distr. of Houses Prices at ocean_proximity={ocean_vals[row, col]}', fontsize=14)
        
        if row!=0:
            plt.xlabel('Houses Prices', fontsize=14)
        if col!=1:
            plt.ylabel('Frequency', fontsize=14)


plt.show()
```
